```python
import pandas as pd

morgan_dataset = pd.read_csv("~/datasets/morgan_dataset.csv")
ap_dataset = pd.read_csv("~/datasets/ap_dataset.csv")
rdk5_dataset = pd.read_csv("~/datasets/rdk5_dataset.csv")
```

```python
morgan_dataset.drop(columns=["pIC50"], inplace=True)
ap_dataset.drop(columns=["pIC50"], inplace=True)
rdk5_dataset.drop(columns=["pIC50"], inplace=True)
```

```python
import os

# Expand `~` to the absolute home directory path
os.makedirs(os.path.expanduser("~/datasets/pycaret"), exist_ok=True)
```

```python
import os
import pandas as pd
from pycaret.classification import *

# File path for saving results
csv_file = "~/datasets/pycaret/morgan_activity.csv"

# Number of trials
n_trials = 40

# Check if the file already exists to determine if we need headers
file_exists = os.path.isfile(csv_file)

# Perform trials
for trial in range(n_trials):
    print(f"Trial {trial + 1}/{n_trials}...")

    # Setup PyCaret
    pyc_morgan_clf = setup(data=morgan_dataset, target='activity', verbose=False)

    # Compare models and select the best one
    best_morgan_clf = compare_models(exclude=["lightgbm"])

    # Get model performance
    model_results = pull()
    model_results["Trial"] = trial + 1  # Add trial number

    # Convert to DataFrame and **append to CSV immediately**
    model_results.to_csv(csv_file, mode='a', header=not file_exists, index=False)
    
    # Update flag so headers aren't written again
    file_exists = True

    print(f"Trial {trial + 1} completed and saved.")

print("Experiment completed. Results continuously saved to 'datasets/pycaret/morgan_activity.csv'.")
```
